```python
#create language translate pyhton appliaction using transformers
from transformers import MarianTokenizer , MarianMTModel
#selecte the model
model_name ='Helsinki-NLP/opus-mt-en-hi'
# load the model using model name
model = MarianMTModel.from_pretrained(model_name)
# load the tokenizer using model name
tokenizer = MarianTokenizer.from_pretrained(model_name)
# get the input from user in text
prompt = input("Enter the text")
# encode the text input into tokens
input_tokens = tokenizer(prompt, return_tensors='pt',padding =True)
# genrate the output token using model
output_token = model.generate(**input_tokens)
#decode the token into final text
print(tokenizer.decode(output_token[0],skip_special_tokens=True))
```
